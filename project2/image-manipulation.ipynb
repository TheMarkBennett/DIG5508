```python
#image Maniplation

from PIL import Image, ImageDraw

filename = "project2\Mark-Bennett.png"
```

```python
# Rotate an image
def roateImg(filename, degrees):

    img = Image.open(filename).rotate(degrees)

    return img

roateImg(filename, 120)
```

```python
# Change an image to grayscale
def grayScale(filename):

    img = Image.open(filename).convert('LA')

    return img

grayScale(filename)
```

```python
# Change the color of an image
def colorChange(filename, filterColor, opacity, grayscale):
    ##get image
    img = Image.open(filename)

    if( grayscale == True):

       img = img.convert('LA')
    
    #Get image size
    width, height = img.size

    #create a new image with the color hex color chosen
    filterImg = Image.new('RGBA', (width, height), filterColor)

    #Create a image mask
    mask = Image.new("L", img.size, opacity)

    #add image mass to mask
    im = Image.composite(img, filterImg, mask)

    return im


colorChange(filename, (255,204,0,255), 50, True )
```

```python
# Change the size of an image

def imageSize(filename, i, j):

    img = Image.open(filename)

    new_img = img.resize((i,j))

    return new_img

imageSize(filename,50,50)
```
